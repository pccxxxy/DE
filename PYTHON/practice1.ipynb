```python
# OPTION D
# The faster
# Only allows processing data row by row without skipping any rows in between 
# Only allows applying the same rule to all rows 

df = pd.read_csv(r'3.4 Wty_Claims_data_new.csv')
df[:20]

def calculate_something(x): # x is the independent variable here
    return float(x) * 1.1

df['total_plus_tax'] = df['Total_Payment'][1:5].map(calculate_something)
df[['total_plus_tax', 'Total_Payment']][:20]
```

```python
calculate_something_new = calculate_something # assign existing function to another function
calculate_something_new(10) # then you can use the functionality of existing function via this new function
```

```python
# OPTION E 
# Suits when you need fucntion to be annoymous
# Use lambda with map or apply to ensure precise capture and full coverage on dataset 

df['total_plus_tax'] = df['Total_Payment'].map(lambda cell: cell * 1.1) # cell is the independent variable here # keyword lambda does the calculation
df[['total_plus_tax', 'Total_Payment']][:20]
```

```python
df['casual part break'] = df['Causal Part Name'].map(lambda x: str(x).lower().split(','))
df['casual part break'] = df['Causal Part Name'].map(lambda x: str(x).lower().split('/'))
df['casual part break'] = df['Causal Part Name'].map(lambda x: str(x).lower().split('+'))
df['casual part first'] = df['Causal Part Name'].map(lambda x: str(x).lower().split(',')[0])

df['casual part break'][:10]
df['casual part first'][:10]
```

```python
# OPTION F
# The faster
# Each individual row and column can be applied with different calculations

df['total_payment_plus_tax'] = df.apply(lambda row: row['Total_Payment'] * 1.2 if row['State'] not in ['QLD', 'NSW', 'VIC'] 
                                        else row['Total_Payment'] * 1.1,
                                        axis = 1)
df['total_payment_plus_tax']
```

# Problem: 
# Detect Fraud in a user's banking transactions

```python
df = pd.read_csv(r'topup_check.csv')
df[:50]
```

Due to humanity nature, fraud(someone else pay with your card) tends to happen when the withdrawl keeps going without top-up within a short period.

How to detect fraud:
<br>
- if the transacation is on the same card
- if the transactions happened in a short period 
- if there is any money goes into your card
- if the card holder has entropy


If the later value in balance is higher than the previous, 
then that indicate there was money going into your card

```python
df['Check_Prev_Bal'] = df['Balance'].shift(1)
df[:5]
```

```python
df['Check_if_Prev_Card'] = df['Card'].shift(1)
df[:5]
```

```python
df['Check_if_Prev_Card'].fillna(0, inplace=True)
df['Check_Prev_Bal'].fillna(0, inplace=True)
df[:5]
```

thinking of the possible factors that can cause the fraud(e.g. normal ways to put money into my card) 
when finding the answer through analysis

```python
df['Topup'] = df.apply(lambda row: True if row['Balance'] > row['Check_Prev_Bal'] 
                       and row['Trans Amt'] > 0  
                       and row['Card'] == row['Check_if_Prev_Card']
                       else False, axis=1)
df
```

```python
import scipy.stats

def calculate_entropy(ary):
    p_data = ary.value_counts()
    entropy = scipy.stats.entropy(p_data)
    
    return entropy 
```
